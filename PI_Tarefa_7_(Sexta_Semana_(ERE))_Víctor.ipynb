#1. Segmentação de Imagens com K-Means

No vídeo abaixo, eu mostro como usar esse algoritmo para fazer quantização de imagens, ou seja, representar as cores originais em menos quantidades.

```
import numpy as np
import cv2
from google.colab.patches import cv2_imshow

def transformaBits(img,bits):
  Z = img.reshape((-1,3))
  Z = np.float32(Z)
  criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 40 , 0.1)
  _,labels, centroides = cv2.kmeans(Z, bits, None, criteria, 40, cv2.KMEANS_RANDOM_CENTERS)
  centroides = np.uint8(centroides)
  imggemColoridaComCentroides = centroides[labels.flatten()]
  imagemFinaI = imggemColoridaComCentroides.reshape((img.shape))
  print("Imagem", bits, 'bits')
  cv2_imshow(imagemFinaI)

img = cv2.imread('/content/drive/MyDrive/Colab Notebooks/vingadores.jpg')
print("Imagem original")
cv2_imshow(img)
transformaBits(img,8)
transformaBits(img,256)
```
